```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
data = load_breast_cancer()
```

```python
X = pd.DataFrame(data['data'], columns=data['feature_names'])
y = data['target'] # malignant, benign
```

```python
X.describe()
```

```python
# visual
import seaborn as sns
import matplotlib.pyplot as plt

# create the count plot
ax = sns.countplot(x=y)
total = len(y)
# Annotate each bar with the percentage
for p in ax.patches:
  percentage = 100 * p.get_height() / total
  ax.text(p.get_x() + p.get_width() / 2, p.get_height(), f'{percentage:.2f}%', ha='center', va='bottom')
plt.xticks([0, 1], ["Malignant", "Benign"])
plt.xlabel("Breast Cancer Diagnostic Distribution")
plt.ylabel("# of Patients")
plt.title("Class Distribution for Breast Cancer Patients")
```

```python
# plot correlation
sns.heatmap(X.corr())
```
